```python
df=pd.read_csv('temp.csv')
df
```

```python
df = df.drop(index=[0])
```

```python
df.attributes=df.attributes.str.split(';')
df.head()
```

```python
def get_parent(attributes): 
    for attribute in attributes: 
        if ('Parent=' in attribute): 
            return attribute
df['parent'] = df.attributes.apply(get_parent)
```
