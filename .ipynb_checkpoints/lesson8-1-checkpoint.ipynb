```python
import pandas as pd
import numpy as np
```

```python
df_1=pd.read_csv(r'F:\Lessons\lesson8\train.csv', sep=',')
df_1
```

```python
df_1[['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude']]=\
np.radians(df_1[['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude']]) #Обращаюсь только к изменяемым колонкам

```

```python
def haversine(pickup_longitude, pickup_latitude, dropoff_longitude, dropoff_latitude, r=6371):
    lat=np.sin((dropoff_latitude-pickup_latitude)/2)
    lon=np.sin((dropoff_longitude-pickup_longitude)/2)
    k=np.cos(dropoff_latitude)*np.cos(pickup_latitude)
    distance=2*r*np.arcsin(np.sqrt(lat**2+k*lon**2))
    return distance
    
```

```python
%%time
df_1['distance'] = pd.Series(haversine(row['pickup_longitude'], row['pickup_latitude'], row['dropoff_longitude'], row['dropoff_latitude']) for i, row in df_1.iterrows())
df_1
```
