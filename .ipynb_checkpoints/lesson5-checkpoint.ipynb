```python
retail_columns=retail.columns
retail_columns
```

```python
retail.drop_duplicates(inplace=True)
retail.shape
```

```python
retail[retail.InvoiceNo.str.startswith('C')] #отбираем только те строки, значение в колонке InvoiceNo у которых начинается с 'C'
```

```python
retail.query('Quantity>0')
```

```python
ret_temp=retail.query('Country=="Germany"').groupby('CustomerID', as_index=False).agg({'InvoiceNo':'nunique'})
tmp=ret_temp.InvoiceNo.quantile(0.8)
```

```python
ret_temp=ret_temp.query('InvoiceNo >@tmp')
```

```python
germany_top=ret_temp["CustomerID"]
```

```python
top_retail_germany=retail.query('CustomerID in @germany_top')

```

```python
top_retail_germany.query('StockCode !="POST"').groupby('StockCode',as_index=False).agg({'Description':'count'}).sort_values("Description", ascending=False)
```
