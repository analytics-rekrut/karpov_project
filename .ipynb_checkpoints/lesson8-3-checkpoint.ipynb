```python
import os
bikes_2=pd.DataFrame()
```

```python
path='F:\\Lessons\\lesson8\\bike_2'
files= os.listdir('F:\\Lessons\\lesson8\\bike_2')
files  
```

```python
for file in files:
        root = path + '\\' +file
        df_start=pd.read_csv(root, sep=',', )
        bikes_2=pd.concat([bikes_2, df_start]).reset_index(drop=True)
bikes_2
```

```python
bikes_2['start_time']=pd.to_datetime(bikes_2['start_time'])
```

```python
bikes_2['end_time']=pd.to_datetime(bikes_2['end_time'])
```

```python
bikes_2.set_index("start_time", inplace=True)
```

```python
bikes_2=bikes_2.groupby('usertype').trip_id.resample(rule = 'D').count()
bikes_2
```

```python
bikes_2=bikes_2.unstack(level = "usertype")
bikes_2
```
