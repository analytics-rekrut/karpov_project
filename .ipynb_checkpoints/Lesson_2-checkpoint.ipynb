```python
import pandas as pd
df=pd.read_csv(r'F:\Lessons\taxi.csv', sep=',')
print(df)
```

```python
df[df.borough =="Brooklyn"].value_counts()
```

```python
df.pickups.sum()
```

```python
df.groupby("borough").agg({"pickups" : "sum"})
```

```python
df.groupby("borough").agg({"pickups" : "sum"}).idxmin()
```

```python
df.groupby(["borough", "hday"]).agg({"pickups" : "mean"})
```

```python
df.groupby(["borough", "pickup_month"], as_index = False).agg({"pickups" : "sum"}).sort_values('pickups', ascending=False)
```
