```python
full_customers=customers.merge(orders, how='inner', on='customer_id')
```

```python
full_customers
```

```python
full_customers.drop(columns=['customer_zip_code_prefix', 'customer_city', 'customer_state', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date'], inplace = True)
```

```python
full_customers['order_approved_at'].dropna(inplace=True) # где нет оплаты того из метрики выкидываем
```

```python
full_customers.drop("year_month", axis=1, inplace=True)
```

```python
full_customers["year_month"]=full_customers['order_approved_at'].dt.date
```

```python
full_customers["year_month"]=pd.to_datetime(full_customers["year_month"])
```

```python
full_customers["year_month"]=full_customers["year_month"].dt.strftime("%Y-%m")
```

```python
full_customers["month"]=full_customers["order_approved_at"].dt.month
```

```python
full_customers["weeks"]=round((full_customers["order_approved_at"].dt.daysinmonth/7), 2) # недели в месяц
```

```python
tpp=full_customers.groupby(["customer_unique_id", "year_month"], as_index=False).agg({'order_purchase_timestamp':'count'})
```
